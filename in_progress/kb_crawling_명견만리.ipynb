```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python
#걸어서 세계속으로 불러오기  
driver = webdriver.Chrome('C://dataitgirls/projects/docu/chromedriver.exe')
driver.get('http://program.kbs.co.kr/1tv/culture/goodinsight/pc/list.html?smenu=c2cc5a')
time.sleep(3)

body = driver.find_element_by_tag_name("body")
```

```python
#영상앨범산
num_of_pagedowns = 100
while num_of_pagedowns:
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(2)
    num_of_pagedowns -= 1
    try:
        driver.find_element_by_css_selecter("#content > div > div.component.ng-scope > div > ul > li> a > span.txt > span.des.ng-binding.ng-scope")#.click()
    except:
        None
        
html = driver.page_source
soup = BeautifulSoup(html, 'lxml')

numbs = soup.find_all("span", "episode ng-binding ng-scope")
dates = soup.find_all("span", "date ng-binding ng-scope")
episodes = soup.find_all("span", "des ng-binding ng-scope")


#content > div > div.component.ng-scope > div > ul > li > a

#content > div > div.component.ng-scope > div > ul > li> a > span.txt > span.des.ng-binding.ng-scope
```

```python
def vod_info(n):
    a = []
    for i in n:
        a.append(i.get_text())
    return a
```

```python
dates = vod_info(dates)
```

```python
dates = pd.DataFrame({'날짜' : dates},
                      columns = {'날짜'})
dates
```

```python
dates.loc[-1] = ['2018.09.28 ']
dates.index = dates.index + 1
dates.sort_index(inplace = True)
dates
```

```python
import re
def vod_info(episodes):
    a = []
    for txt in episodes:
        text = re.sub('<[^<]+?>', ' ', str(txt))
        a.append(text)
    return a
```

```python
episodes = vod_info(episodes)
episodes
```

```python
episodes = pd.DataFrame({'시놉시스' : episodes},
                      columns = {'시놉시스'})
episodes
```
